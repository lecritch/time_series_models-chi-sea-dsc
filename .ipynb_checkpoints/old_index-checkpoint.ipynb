# Time Series Modeling

Now that we have learned how to prepare our data for modeling, let's get to the actual models.

We have adjusted our data to include a datetime object as index, removed extreme values and interpolated the resulting na's, and differenced our data to make it stationary.


If there were no patterns left in our data, our sample would simply be what is called noise. 

It would randomly fluctuate around the mean.

$$\Large Y_t =  \epsilon_t$$

```python
# Let's make some white noise!
from random import gauss as gs
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

rands = []
for _ in range(1000):
    rands.append(gs(0, 1))
series = pd.Series(rands)
```

```python
X = np.linspace(-10, 10, 1000)
plt.figure(figsize=(10, 7))
plt.plot(X, series);
```

```python
import statsmodels.api as sm
time_index = pd.to_datetime(series.index, unit='D')
series.index = time_index
decomp_white = sm.tsa.seasonal_decompose(series)
```

```python
plt.figure(figsize=(20, 8))
plt.subplot(411)
plt.plot(series.index, decomp_white.observed, label='Original')
plt.legend()
plt.subplot(412)
plt.plot(series.index, decomp_white.trend, label='Trend')
plt.legend()
plt.subplot(413)
plt.plot(series.index, decomp_white.seasonal, label='Seasonal')
plt.legend()
plt.subplot(414)
plt.plot(series.index, decomp_white.resid, label='Residual')
plt.legend()
plt.subplots_adjust(hspace=0.5);
```

One step up from a white noise model is a random walk.

$$\Large Y_t = Y_{t-1} + \epsilon_t$$

Let's bring back our Chicago gun crime data

```python
ts = pd.read_csv('data/Gun_Crimes_Heat_Map.csv')
ts['Date'] = pd.to_datetime(ts.Date)
ts_minute = ts.groupby('Date').count()['ID']
daily_count = ts_minute.resample('D').sum()
daily_count = daily_count[daily_count < 90]
```

```python
ts_dr = pd.date_range(daily_count.index[0], daily_count.index[-1])
ts_daily = np.empty(shape=len(ts_dr))
ts_daily = pd.Series(ts_daily)
ts_daily = ts_daily.reindex(ts_dr)
ts_daily = ts_daily.fillna(daily_count)
ts_daily = ts_daily.interpolate()
```

A random walk model for our timeseries would simply predict the previous value for the current value

```python
# we can perform this with the shift operator

random_walk = ts_daily.shift(1)
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()

ts_daily[0:10].plot(ax=ax, c='r', label='original')
random_walk[0:10].plot(ax=ax, c='b', label='shifted')
plt.legend()
```

```python
ts_daily.diff()
```

Let's look at the RMSE if we just predicted the previous value.

```python
ts_daily
```

```python
residuals = random_walk - ts_daily
np.sqrt((residuals.dropna()**2).sum())
```

# Autoregressive Models

An autoregressive model is a model that a timeseries regressed on its previous values

### $y_{t} = c + \phi_{1}y_{t-1} + \varepsilon_{t}$

The above formula is a first order AR model, aka AR(1). It uses the previous point to calculate the next point.  

In code, we create the model with:  
  

```python
from statsmodels.tsa.ar_model import AR,ARResults
```

```python
model = AR(ts_daily)
AR1fit = model.fit(maxlag=1,method='mle')
predictions1 = AR1fit.predict(dynamic=False ).rename('AR(1) Predictions')
AR1fit.summary()
```

Let's also fit a linear regression to the lagged data

```python
from sklearn.linear_model import LinearRegression


lr = LinearRegression()
lr.fit(np.array(ts_daily.shift(1).dropna()).reshape(-1,1), ts_daily[1:])
print(lr.coef_)
print(lr.intercept_)
```

Our coefficients are the same.

We don't have to limit ourselves to one lag either.  We can go back as far as we like.

 $$\large y_{t} = c + \phi_{1}y_{t-1} + \phi_{2}y_{t-2} + \dots + \phi_{p}y_{t-p} + \varepsilon_{t}$$

```python
model = AR(ts_daily)
AR2fit = model.fit(maxlag=2,method='mle')
predictions1 = AR1fit.predict(dynamic=False ).rename('AR(1) Predictions')
AR2fit.summary()
```

```python
s_1 = pd.Series(ts_daily.shift(2).dropna()),
s_2 = pd.Series(ts_daily.shift(1).dropna())
shift = pd.DataFrame(s_1).T

shift['s_2'] = s_2
shift.columns = ['s_1', 's_2']
shift.head()
```

```python
lr = LinearRegression()
lr.fit(shift, ts_daily[2:])
print(lr.coef_)
print(lr.intercept_)
```

Let's look at our RMSE

```python
np.sqrt(sum(AR2fit.resid**2))
```

```python
Our RMSE improved.
```

# Moving Average Model

A moving average model uses the errors from previous forcasts to predict future points.

$$\large Y_t = \mu +\epsilon_t + \theta * \epsilon_{t-1}$$

Just like with AR models, we can fit MA models as many steps back as we like.  

The above formula refers to a first order MA model MA(1)

```python
from statsmodels.tsa.arima_model import ARMA
```

```python
# use 1 in the 2nd position in the ARMA order parameter
ma = ARMA(ts_daily, (0,1)).fit(maxlag=1)
predictionsma = ma.predict(dynamic=False ).rename('AR(1) Predictions')
predictionsma
```

```python
ma.summary()
```

```python
To reproduce these results, we can calculate our errors 
```

```python
manual_preds = (ma.params['const'] + ma.resid*ma.params['ma.L1.y']).shift(1)
manual_preds
```

```python
np.sqrt((ma.resid**2).sum())
```

```python
If we plot the residuals, however, we see the following
```

```python
plt.plot(ma.resid.index, ma.resid)
```

```python
plt.plot(ma.resid.index, ma.resid)
plt.plot(ma.resid.index, ma.resid.rolling(365).mean())
plt.plot(ma.resid.index, ma.resid.rolling(365).std())
```

The residuals mirror the shape of the real data. We can see that the model is on average more accurate after 2017, when the trend levels off.

Remember that we said before that our models required our data to be stationary.  We fed the non-stationary data, so our model varied in accuracy across time

# Integrated

The second spot in the order of the ARIMA model (d) represents the integrated component.  Integrated is another way to refer to differencing.  If we put 1 in the d spot, we have one level of differencing.  We found from our Dickey-Fuller test that 1 order of differencing made our models stationary, so let's try it out.

```python
from statsmodels.tsa.arima_model import ARIMA
```

```python
arima = ARIMA(ts_daily, (0,1,1)).fit()
y_hat = arima.predict(typ='levels')
```

```python
arima.summary()
```

Let's compute our RMSE

```python
# Our best so far
np.sqrt((arima.resid**2).sum())
```

Let's look again at our residual plot

```python
plt.plot(arima.resid.index, arima.resid)
plt.plot(arima.resid.index, arima.resid.rolling(365).std())
```

This helps us see why our model is called moving average.  The errors oscillate consistantly around the mean of zero.

## Auto-regressive Time Series Model

An autoregression model makes an assumption that the observations at previous time steps are useful to predict the value at the next time step. It is one of the simplest time series models in which we use a linear model to predict the value at the present time using the value at the previous time. 

<p style='text-align:center; font-size: 30px;'><sub>t</sub>=<sub>1</sub><sub>1</sub>+<sub></sub></p>

The numeral one (1) denotes that the next instance is solely dependent on the previous instance.  The (phi) is a coefficient which we seek so as to minimize the error function.

The order of AR is the number of lag terms we are using to predict the present value (AR(1) uses only 1 lag - one value directly preceding the value you are trying to predict, AR(2) use the two values directly preceding the value you are trying to predict) 

#### How do we determine the order aka how many lag terms do we include? 

Using ACF and PACF! 

<img src='img/AR(1).png'>

```python
from statsmodels.tsa.arima_model import ARIMA
#fit ARIMA model (1,1,0))

model = ARIMA(ts_daily, order=(1,1,0))
model_fit = model.fit(disp=0)
print(model_fit.summary())
```

## Moving Average Time Series Model

Sometimes, a past value is not a useful indicator of what value will come next. Consider a system that is subject to a lot of shocks/volatility. If a previous time period experiences a shock it may cause an error for future predictions if we just that value. A moving average model helps address this behavior. 

A moving average term in a time series model is a past error (multiplied by a coefficient).

An MA model assumes present value is related to errors in the past - includes memory of past errors


<p style='text-align: center; font-size:30px;'><sub>t</sub>= + <sub></sub>+<sub>1</sub><sub>1</sub></p>
